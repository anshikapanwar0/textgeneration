```python
#Tokenization
import nltk
from nltk import word_tokenize
nltk.download('punkt')
data['sentence_tokens']= data['Sentence'].apply(word_tokenize)
data['target_tokens']=data['Target'].astype(str).apply(word_tokenize)
data.head()
```

```python
# Removing punctuations
def remove_punct(tokens):
    return [word for word in tokens if word.isalpha()]

data['sentence_tokens_clean'] = data['sentence_tokens'].apply(remove_punct)
data['target_tokens_clean'] = data['target_tokens'].apply(remove_punct)
print(data[['Sentence', 'Target', 'sentence_tokens', 'target_tokens', 'sentence_tokens_clean', 'target_tokens_clean']])
```

```python
#Stemming
from nltk.stem import PorterStemmer
ps= PorterStemmer()
data['ps_stem_sent']=data['sentence_tokens_clean'].apply(lambda x: [ps.stem(word) for word in x])
data['ps_stem_target']=data['target_tokens_clean'].apply(lambda x: [ps.stem(word) for word in x])
data.head()
```

```python
#lemmetization
nltk.download('wordnet')
from nltk.stem.wordnet import WordNetLemmatizer
lemmatizer= WordNetLemmatizer()
data['ps_lemm_sent']=data['sentence_tokens_clean'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
data['ps_lemm_target']=data['target_tokens_clean'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
data.head()
```

```python
#POS Tagging
nltk.download('averaged_perceptron_tagger')
data['pos_tag_sent']=data['ps_lemm_sent'].apply(nltk.pos_tag)
data['pos_tag_target']=data['ps_lemm_target'].apply(nltk.pos_tag)
data.head()
```

```python
#Model Architecture
#Training word2vec with gensim
from gensim.models import Word2Vec
sent_model = Word2Vec(sentences=data['ps_lemm_sent'], vector_size=100, window=5, min_count=1, workers=4)
target_model = Word2Vec(sentences=['ps_lemm_target'], vector_size=100, window=5, min_count=1, workers=4)
sent_model.save("sent_word2vec.model")
target_model.save("target_word2vec.model")
sent_model = Word2Vec.load("sent_word2vec.model")
target_model = Word2Vec.load("target_word2vec.model")

def tokens_to_vectors(tokens, model):
    return [model.wv[word] for word in tokens if word in model.wv]
data['sent_vectors'] = data['ps_lemm_sent'].apply(lambda tokens: tokens_to_vectors(tokens, sent_model))
data['target_vectors'] = data['ps_lemm_target'].apply(lambda tokens: tokens_to_vectors(tokens, target_model))
print(data[['ps_lemm_sent', 'sent_vectors', 'ps_lemm_target', 'target_vectors']].head())
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

# --- Data Extraction ---
# Access the columns of the DataFrame using numerical indices
sent_data = data['ps_lemm_sent']
target_data = data['ps_lemm_target']

tokenizer = Tokenizer()
tokenizer.fit_on_texts(sent_data)  # Fit on the sentence data
vocab_size = len(tokenizer.word_index) + 1

# --- Sequence Conversion and Padding ---
sent_sequences = tokenizer.texts_to_sequences(sent_data)
target_sequences = tokenizer.texts_to_sequences(target_data)
max_seq_length = max(max([len(seq) for seq in sent_sequences]), max([len(seq) for seq in target_sequences]))
sent_sequences = pad_sequences(sent_sequences, maxlen=max_seq_length, padding='post')
target_sequences = pad_sequences(target_sequences, maxlen=max_seq_length, padding='post')

# --- Data Splitting ---
X_train, X_val, y_train, y_val = train_test_split(sent_sequences, target_sequences, test_size=0.5, random_state=42)

print("Training set size (X_train, y_train):", X_train.shape, y_train.shape)
print("Validation set size (X_val, y_val):", X_val.shape, y_val.shape)
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Embedding, GRU, Dense, Dropout
from keras.regularizers import l2

embedding_dim = 100

model = Sequential()
model.add(Embedding(input_dim=vocab_size,
                    output_dim=embedding_dim,
                    input_length=max_seq_length))
model.add(GRU(units=128,
              return_sequences=True,
              kernel_regularizer=l2(0.01)))
model.add(Dropout(0.2))
model.add(Dense(units=vocab_size, activation='softmax'))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Reshape targets to fit the model output
y_train = np.expand_dims(y_train, -1)
y_val = np.expand_dims(y_val, -1)

# Train the model
model.fit(X_train, y_train, epochs=20, validation_data=(X_val, y_val), batch_size=32)

# Save the model
model.save('model.h5')
```

```python
import numpy as np
from keras.preprocessing.sequence import pad_sequences

def generate_text(seed_text, model, tokenizer, max_seq_length, temperature=0.7):
    seed_sequence = tokenizer.texts_to_sequences([seed_text])
    seed_sequence = pad_sequences(seed_sequence, maxlen=max_seq_length, padding='post')

    generated_text = seed_text

    for _ in range(max_seq_length):  # Adjust the number of words to generate as needed
        predictions = model.predict(seed_sequence, verbose=0)[0]
        predictions = predictions / temperature
        predicted_id = np.argmax(predictions[-1, :])
        predicted_word = tokenizer.index_word.get(predicted_id, '')

        if predicted_word == '':  # Stop if the word is not in the vocabulary
            break

        generated_text += ' ' + predicted_word

        seed_sequence = np.append(seed_sequence[:,1:], [[predicted_id]], axis=1)

        if predicted_word == 'end':  # Stop if the end token is generated
            break

    return generated_text

```
